```python
def get_region(link):
    if "World" in link:
        return "WORLDS"
    elif "LEC" in link:
        return "LEC"
    elif "LCK" in link:
        return "LCK"
    elif "LPL" in link:
        return "LPL"
    elif "LCS" in link:
        return "LCS"
    elif "MSI" in link:
        return "MSI"
    else:
        return "Other"

tournaments2023_tuples = [(link, get_region(link), 2023) for link in tournaments2023]
tournaments2022_tuples = [(link, get_region(link), 2022) for link in tournaments2022]
tournaments2021_tuples = [(link, get_region(link), 2021) for link in tournaments2021]
tournaments2020_tuples = [(link, get_region(link), 2020) for link in tournaments2020]
tournaments2019_tuples = [(link, get_region(link), 2019) for link in tournaments2019]

tournaments = tournaments2023_tuples + tournaments2022_tuples + tournaments2021_tuples + tournaments2020_tuples

display(tournaments)
```

```python
def get_matches(tournament_url, region, year):
    page = retrieve.firefoxPage(tournament_url)
    soup = BeautifulSoup(page, 'lxml')
    match_list = soup.find('table').find('tbody')
    matches = []
    for match in match_list:
        try : suffix = match.find('a').get('href')
        except :
            display(match) # Apparently it's the match betwen GEN and BRO in 2022 where GIGABRO FF'd so there's no link / details
        # Checking if it is really a match suffix by checking if it is in the format like : "../game/stats/26935/page-summary/"
        if "../game/stats/" in suffix:
            suffix = suffix[2:] # Removing the first two characters which are ".."
            match_link = "https://gol.gg" + suffix
            matches.append((match_link, region, year))
    return matches

# print(get_matches(tournaments[0][0], tournaments[0][1]))

def get_all_matches(tournaments):
    matches = []
    for tournament in tournaments:
        matches.extend(get_matches(tournament[0], tournament[1], tournament[2]))
    return matches

matches = get_all_matches(tournaments)
display(matches)
```

```python
def get_games(match_url, region, year):
    page = retrieve.firefoxPage(match_url)
    soup = BeautifulSoup(page, 'lxml')
    game_links = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and "/game/stats/" in href and "/page-game/" in href:
            suffix = href[2:]
            link = "https://gol.gg" + suffix
            game_links.append((link, region, year))
    return game_links

# print(matches[0][0])
# print(get_games(matches[0][0], matches[0][1], matches[0][2]))

def get_all_games(matches):
    games = []
    for match in matches:
        games.extend(get_games(match[0], match[1], match[2]))
    return games

games = get_all_games(matches)
```

```python
def get_game_result(game_url, region, year):
    page = retrieve.firefoxPage(game_url)
    soup = BeautifulSoup(page, 'lxml')

    blueside = soup.find_all('div', class_="col-12 blue-line-header")
    redside = soup.find_all('div', class_="col-12 red-line-header")
    game_duration = soup.find('div', class_="col-6 text-center").find('h1').text

    blueside_team = blueside[0].find('a').text
    redside_team = redside[0].find('a').text
    blueside_victory = "WIN" in blueside[0].text
    redside_victory = "WIN" in redside[0].text

    return (blueside_team, blueside_victory, redside_team, redside_victory, game_duration, region, year)

get_game_result(df["Link"][0], df["Region"][0], df["Year"][0])
```

```python
def get_game_stats(game_url, region, year, blue_team, red_team, blue_win, red_win, duration):
    # replacing page-game/ suffix into page-fullstats/ in the game_url
    game_url = game_url.replace("page-game/", "page-fullstats/")
    page = retrieve.firefoxPage(game_url)
    soup = BeautifulSoup(page, 'lxml')
    label = ["Side", "Region", "Year", "Game ID", "Team", "Victory", "Game Duration"]
    initial_size = len(label)
    print(game_url)
    # https://gol.gg/game/stats/57150/page-fullstats/, getting the id between stats/ and /*****
    game_id = game_url.split("stats/")[1].split("/")[0]
    player_data = []
    for i in range(5):
        player_data.append(["Blue Side", region, year, game_id, blue_team, blue_win, duration])
    for i in range(5):
        player_data.append(["Red Side", region, year, game_id, red_team, red_win, duration])
    table = soup.find_all('tr')
    for row in table:
        td = row.find('td')
        if not td:
            continue
        label.append(td.text)
        for i, cell in enumerate(row.find_all('td')[1:]):
            player_data[i].append(cell.text)
    if (len(label) != len(player_data[0])):
        return None
    return pd.DataFrame(player_data, columns=label)

def get_game_full(game_url, region, year):
    blue_team, blue_win, red_team, red_win, duration, region, year = get_game_result(game_url, region, year)
    return get_game_stats(game_url, region, year, blue_team, red_team, blue_win, red_win, duration)
    
print(get_game_full(df["Link"][0], df["Region"][0], df["Year"][0]))
```

```python
def get_all_game_full(df):
    for i in range(len(df)):
        if i == 0:
            full_data = get_game_full(df["Link"][i], df["Region"][i], df["Year"][i])
        else:
            res = get_game_full(df["Link"][i], df["Region"][i], df["Year"][i])
            if res is not None:
                full_data = pd.concat([full_data, res])
    return full_data

full_data = get_all_game_full(df)
```
